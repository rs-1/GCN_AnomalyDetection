```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
```

```python
df = pd.read_csv('aucs.csv')
print(df)
```

```python
by_source = df.set_index('Alpha').groupby('Source')
by_source.describe()
```

```python
fig, ax = plt.subplots()
by_source['AUC'].plot(ax=ax, linestyle='-', marker='D')
plt.xlabel('Network    <====    Alpha    ====>    Attributes')
plt.ylabel('AUC')
plt.title('Dataset AUCs vs Alpha (attribute/network weighting)')
plt.legend(bbox_to_anchor=(1, 1))
plt.show()
```

```python
for k, v in by_source:
    print(by_source.get_group(k), "\n\n")
```
